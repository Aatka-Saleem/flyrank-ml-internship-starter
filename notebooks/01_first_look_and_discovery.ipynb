## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

## 3. 🔧 Your turn

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

- Redo Discovery A but only for pages with `impressions_90d > 0` — does the correlation change?
- In Discovery B, which `content_type` has the worst CTR *at the same position tier*?
- Find another belief to test: does `content_age_days` relate to `trend_direction`? Does `avg_position` relate to `ctr`?

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
# My discovery here
# OPTION 1: Does correlation change when we filter out pages with zero impressions?
visible_impressions = df[df["impressions_90d"] > 0]
corr_filtered = visible_impressions["search_volume"].corr(visible_impressions["impressions_90d"])
corr_all = df["search_volume"].corr(df["impressions_90d"])

print(f"Correlation (all pages):                    {corr_all:.3f}")
print(f"Correlation (pages with impressions > 0):  {corr_filtered:.3f}")
print(f"\nObservation: filtering pages with actual traffic shows {abs(corr_filtered):.3f}")
print("→ Even pages that got impressions show weak correlation between search volume and traffic.")
```

```python
# OPTION 2: Which content type underperforms at the same position?
visible = df[df["impressions_90d"] >= 100]

# Compare CTR by content_type WITHIN a specific position tier
# Let's use "page_1" (top 10 position)
page_1_content = visible[visible["position_tier"] == "page_1"]
ctr_by_type = page_1_content.groupby("content_type")["ctr"].agg(["mean", "count"])
ctr_by_type = ctr_by_type.sort_values("mean", ascending=True)

print("CTR by content type (position_tier = 'page_1', impressions >= 100):")
print(ctr_by_type.round(4).to_string())
print(f"\nObservation: {ctr_by_type.index[0]} has the lowest mean CTR at position 1-10")
print("→ At the same search position, some content types convert visitors less effectively.")
```

```python
# OPTION 3: Does older content decline more than newer content?
age_by_trend = df.groupby("trend_direction")["content_age_days"].median()
print("Median content age by trend direction:")
print(age_by_trend.round(0).to_string())

# Also check the distribution visually
import matplotlib.pyplot as plt
df[df["content_age_days"] > 0].boxplot(column="content_age_days", by="trend_direction")
plt.title("Distribution of content age by trend direction")
plt.ylabel("Content age (days)")
plt.suptitle("")
plt.show()

print("\nObservation: pages trending 'down' vs 'up' have similar age distributions")
print("→ Content age alone does not explain which pages are declining.")
```

```python
# OPTION 4: Better position = higher CTR?
# Filter out "no_data" position rows
positional = df[df["avg_position"] > 0]
corr_pos_ctr = positional["avg_position"].corr(positional["ctr"])

print(f"Correlation between avg_position and CTR: {corr_pos_ctr:.3f}")
print(f"\nInterpretation: correlation of {corr_pos_ctr:.3f} means...")
if corr_pos_ctr < -0.5:
    print("  → Strong negative: being higher in results DOES lead to better CTR")
elif corr_pos_ctr > -0.2:
    print("  → Weak to no correlation: position alone doesn't determine CTR")
else:
    print("  → Moderate: position matters, but other factors play a role")
```
