**I am choosing Lane 2:**

 Refresh / Content Opportunity Scoring. In any large-scale digital content inventory, pages naturally lose search visibility, decay in freshness, or experience shifts in user search behavior over time. Identifying these decaying pages manually across thousands of URLs is operationally impossible. This lane provides a structured framework to build a predictive machine learning pipeline that surfaces high-priority review candidates before their traffic drops completely away, maximizing content lifecycle value and optimization efficiency.


## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**The Core Question:**

Which high-visibility or high-potential web pages are showing active decay, underperformance, or freshness risks, and should be prioritized for immediate editorial review?

**The Unit of Analysis (Grain):**
  
A single pseudonymized content item (web page).  

**The Decision and Action:**

This work directly improves the content allocation decision. Instead of guessing or choosing pages at random, the content management team will use the ranked output to take specific actions: updating outdated data, expanding thin articles, rewriting titles/meta fields, or merging overlapping topics.

**The Cost of a Wrong**

Call:False Positive (Flagging a healthy page): Wastes expensive human editorial hours rewriting an article that didn't need it, potentially breaking its stable search standing.

False Negative (Missing a decaying page): Leads to a permanent loss of search visibility and organic user traffic to competitors, dragging down revenue.

**Why ML is Necessary:**

While a manual SQL rule can flag basic metrics like "has not been updated in 180 days," it treats all pages uniformly. A machine learning ranking model can ingest multiple non-linear engagement, visibility, and trend signals simultaneously, learning complex probability curves to rank the review queue with far higher precision.

Before designing complex models, we must verify that the underlying data exhibits the variance and indicators required to justify a refresh prioritization workflow. Below, I am loading the starter playground dataset to extract foundational baseline metrics regarding content age, current trend distribution, and potential baseline metrics.

```python
import os
import sys
import subprocess
import pandas as pd

# 1. Environment Detection and Repository Setup
IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/Aatka-Saleem/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
else:
    # Navigate up to repository root if launched from notebooks/ or work/notebooks/
    while not os.path.exists("data/raw/content_refresh_anonymized.csv") and os.path.dirname(os.getcwd()) != os.getcwd():
        os.chdir("..")

print("Working dir:", os.getcwd())

# 2. Path Validation
data_path = "data/raw/content_refresh_anonymized.csv"
assert os.path.exists(data_path), "Starter CSV not found — are you at the repo root?"
print("Starter data found. You're ready.")

# 3. Load Dataset and Compute Lane 2 Metrics
print("\nLoading dataset to parse supporting numbers...")
df = pd.read_csv(data_path)

# Total volume & scope
total_pages = len(df)

# Percentage of pages actively declining (The Target Proxy)
declining_count = (df['trend_direction'] == 'down').sum()
pct_declining = (declining_count / total_pages) * 100

# Average content age & exposure metrics
avg_age = df['content_age_days'].mean()
high_impact_stale = df[(df['content_age_days'] >= 180) & (df['impressions_90d'] >= 500)].shape[0]

# Print out your real-world supporting evidence
print(f"\n" + "="*30)
print(f"--- LANE 2: DATASET CORE METRICS ---")
print(f"="*30)
print(f"Total Unique Content Items: {total_pages:,}")
print(f"Actively Declining Pages:   {declining_count:,} ({pct_declining:.2f}%)")
print(f"Average Inventory Age:      {avg_age:.1f} days")
print(f"High-Visibility Stale Pages (>=180d age & >=500 impressions): {high_impact_stale:,}")
print(f"="*30)
```
